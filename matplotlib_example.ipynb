### Matplotlib

Veri görselleştirme kütüphanesidir. Temel olarak 2D grafikleri çizmek için kullanılır. (Bazı ek pluginlerle 3D grafikler oluşturulabilir.)

```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4,5]
y=[10,20,30,40,50]
```

```python
x,y
```

### Grafik 1 İsterleri

- Çizgi rengini yeşil yapınız.
- X ekseni değerleri 0'dan başlayacak şekilde güncelleyiniz. (Arraydaki ilk değer 0 olmasın)

```python
plt.plot(x,y, marker="s", linestyle="dotted", color="red", label="Satışlar")
plt.xlabel("Günler")
plt.ylabel("Satışlar")
plt.title("Günlük Satışlar")
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.plot(x,y, marker="s", linestyle="dotted", color="green", label="Satışlar")
plt.xlabel("Günler")
plt.ylabel("Satışlar")
plt.title("Günlük Satışlar")
plt.legend()
plt.xlim(0)
plt.grid(True)
plt.show()
```

### Çoklu Grafik

İsterler:
- Yeni bir grafikte tan cot grafiği

```python
import numpy as np
```

```python
x = np.linspace(1,10,100)
x
```

```python
y1 = np.sin(x)
y2 = np.cos(x)
```

```python
y1, y2
```

```python
plt.figure(figsize=(8,6))
plt.plot(x,y1, label="sin(x)", color="red")
plt.plot(x,y2, label="cos(x)", color="blue")
plt.legend()
plt.xlabel("X Değeri")
plt.ylabel("Fonksiyonun Değeri (sin,cos)")
plt.grid(True)
```

```python
tan_y = np.tan(x)
cot_y = 1 / np.tan(x)
```

```python
plt.figure(figsize=(8,6))
plt.plot(x,tan_y, label="tan(x)", color="red")
plt.plot(x,cot_y, label="cot(x)", color="blue")
plt.legend()
plt.xlabel("X Değeri")
plt.ylabel("Fonksiyonun Değeri (tan,cot)")
plt.grid(True)
```

```python
x = np.linspace(-2*np.pi, 2*np.pi, 1000)
y1 = np.tan(x)
y2 = 1 / np.tan(x)

y1[np.abs(y1) > 10] = np.nan
y2[np.abs(y2) > 10] = np.nan

plt.figure(figsize=(8, 6))
plt.plot(x, y1, label="tan(x)", color="red")
plt.plot(x, y2, label="cot(x)", color="blue")
plt.legend()
plt.xlabel("x değeri")
plt.ylabel("Fonksiyon Değeri")
plt.title("Tan ve Cot Grafiği")
plt.grid(True)
plt.show()
```

### Bar Grafiği

İsteri yok

```python
categories = ["A", "B", "C","D","E","F"]
values = [5,10,20,40,50,30]
```

```python
plt.bar(categories, values, color=["red","blue","green","purple","orange","black"])
plt.xlabel("Kategoriler")
plt.ylabel("Değerler")
plt.title("Kategori bazlı satışlar")
```
